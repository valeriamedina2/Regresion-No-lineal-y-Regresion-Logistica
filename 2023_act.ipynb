```python
#Cargamos librerias
import pandas as pd

import numpy as np
import matplotlib.pyplot as plt
import scipy.special as special
from scipy.optimize import curve_fit
import seaborn as sns
from sklearn.metrics import r2_score
```

```python
#Cargar 
df= pd.read_csv('Gastos y costos 2023_sin_nulos.csv')
df

```

```python
#Graficamos todas las dispersiones entre todas las variables
sns.pairplot(df)
```

```python
#Declaramos las variables dependientes e independientes para la regresin No lineal
Vars_Indep= df[['IVA']]
Var_Dep= df['TOTAL MX']
```

```python
#Redefinimos las variables 
x= Vars_Indep
y= Var_Dep
```

   ** Usamos las siguientes ecuaciones como funciones de ajuste.**

            y = ax^2 + bx + c

            y = ax^3 + bx + c

            y = ax^3 + bx^2 + c

            y = a*exp(bx) + c

**MODELO 1**

```python
def  func1 (x, a, b, c):
     return a*x**2 + b*x + c
```

```python
#Ajustamos los parmetros de la funcin curve_fit
parametros, covs= curve_fit(func1, df['IVA'], df['TOTAL MX'])
```

```python
#Obtenemos los coeficientes del modelo de regresin no lineal
parametros
```

```python
#Creamos el modelo de prediccin con los parmetros obtenidos 
parametros, _ = curve_fit(func1, df['IVA'], df['TOTAL MX'])
a, b, c = parametros[ 0 ], parametros[ 1 ], parametros[ 2 ]
yfit1 = a * x ** 2 + b * x + c
```

```python
#Calculamos las predicciones y reestructuramos el vector de predicciones
#x= x.reshape(51, 1)
#y = y.reshape(51, 1)
#yfit1 = yfit1.reshape(51, 1)
yfit1
```

```python
#Graficamos las predicciones y los datos originales para realizar la comparacin
plt.plot(x, y, 'bo', label="y-original")
plt.plot(x, yfit1, label="y=a*x^2+b*x+c")
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='best', fancybox=True, shadow=True)
plt.grid(True)
plt.show() 
```

```python
#Calculamos el coeficiente de determinacin del modelo

R2 = r2_score(y, yfit1) 
R2
```

```python
#Imprimimos el scatter plot entre la variable dependiente (total) e independiente (alchool) 
#para observar el comportamiento en su dispersin 
from turtle import color


sns.scatterplot(x='IVA', y='TOTAL MX', color="blue", data=df)
#sns.scatterplot(X='speeding', y='total', color="red", data=df)
```

```python
#Declaramos las variables dependientes e independientes para la regresin lineal
#Vars_Indep= df[['alcohol', 'speeding']]
Vars_Indep= df[['IVA']]
Var_Dep= df['TOTAL MX']
```

```python
#Se define model como la funcin de regresin lineal
from sklearn.linear_model import LinearRegression
model= LinearRegression()
```

```python
#Verificamos la funcin relacionada al modelo
type(model)
```

```python
#Ajustamos el modelo con las variables antes declaradas
model.fit(X=Vars_Indep, y=Var_Dep)
```

```python
#Verificamos los coeficientes obtenidos para el modelo ajustado
model.__dict__
```

```python
#Evaluamos la eficiencia del modelo obtenido por medio del coeficiente R Determinacin
model.score(Vars_Indep,Var_Dep)
```

```python
#Predecimos los valores de total de accidentes a partir de la variable "alcohol"
#y_pred= model.predict(X=df[['alcohol', 'speeding']])
y_pred= model.predict(X=df[['IVA']])
y_pred
```

```python
#Insertamos la columna de predicciones en el DataFrame
df.insert(0, 'Predicciones', y_pred)
df
```

```python
#Corroboramos cual es el coeficiente de Determinacin de nuestro modelo
coef_Deter=model.score(X=Vars_Indep, y=Var_Dep)
coef_Deter
```

```python
#Corroboramos cual es el coeficiente de Correlacin de nuestro modelo
coef_Correl=np.sqrt(coef_Deter)
coef_Correl
```
